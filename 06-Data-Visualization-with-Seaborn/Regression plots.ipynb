```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.lmplot(x='total_bill', y='tip',data=tips,hue='sex',markers=['o','v'])
```

```python
sns.lmplot(x='total_bill', y='tip',data=tips,col='day',hue='time')
```

```python
sns.lmplot(x='total_bill', y='tip',data=tips,col='day',hue='time',aspect=0.5,height=8)
```
