```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['tip'],kde=False, bins=10)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.pairplot(tips,hue='sex',palette='cool')
```

```python
sns.rugplot(tips['total_bill'])
```
